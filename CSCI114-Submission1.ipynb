```python
import pandas as pd
```

# Wine

```python
file_wine_data = "wine.csv"

wine_data = pd.read_csv(file_wine_data)

wine_data
```

```python
# Labels
x_wine = wine_data.iloc[:,:11]

x_wine
```

```python
# How many features do we have?

wine_num_features = len(x_wine.columns)

print("Number of Features :{}".format(wine_num_features))
```

```python
wine_columns = []

# Label all x-features
for i in range(wine_num_features):
    wine_columns.append("x{}".format(i))
    
x_wine.columns = wine_columns

x_wine
```

```python
x_wine_mean = x_wine.mean()
x_wine_std = x_wine.mean()
x_wine_standardized = (x_wine - x_wine_mean)/x_wine_std

x_wine_standardized
```

```python
# Binary Classification
y_wine = wine_data['quality'].replace(['bad'], 0).replace(['good'], 1)

y_wine = y_wine.values

y_wine
```

```python
num_bad = len(wine_data[wine_data.iloc[:, 11] == 'bad'])
num_good = len(wine_data[wine_data.iloc[:, 11] == 'good'])

print("num_bad: {}".format(num_bad))
print ("num_good: {}".format(num_good))
```

```python
df_wine = x_wine_standardized.copy()
df_wine['y_wine'] = y_wine
# adding or creating a new column
df_wine
```

```python
def partition_dataset(df, num_a=20, num_b=20, val_a=1, val_b=0):
    df_a = df[df.iloc[:,-1] == val_a].sample(num_a)
    df_b = df[df.iloc[:,-1] == val_b].sample(num_b)
    
    #remove it
    df.drop(df_a.index, inplace=True)
    df.drop(df_b.index, inplace=True)
    
    # samples: df_a and df_b
    # simply combining two
    frames = [df_a, df_b]
    df_validation = pd.concat(frames)
    
    return df, df_validation

wine_training, wine_validation = partition_dataset(df_wine)
```

```python
wine_training
```

```python
wine_validation
```
