```python
import pandas as pd

file_raw_data = "wine.csv"

raw_data = pd.read_csv(file_raw_data)

raw_data
```

```python
x = raw_data.iloc[:,:11]

x
```

```python
num_features = len(x.columns)

print("Number of Features: {}".format(num_features))
```

```python
columns = []

for i in range(num_features):
    columns.append("x{}".format(i))
    
x.columns = columns

x
```

```python
x_normalized = (x - x.min()) / (x.max() - x.min())

x_normalized
```

```python
y = raw_data['quality'].replace(['bad'], 0).replace(['good'], 1)

y = y.values

y
```

```python
num_bad = len(raw_data[raw_data.iloc[:, 11] == 'bad'])
num_good = len(raw_data[raw_data.iloc[:, 11] == 'good'])

print("num_bad: {}".format(num_bad))
print ("num_good: {}".format(num_good))
```

```python
df = x_normalized.copy()
df['y'] = y
# adding or creating a new column
df
```

```python
def partition_dataset(df, num_a=50, num_b=50, val_a=1, val_b=0):
    df_a = df[df.iloc[:,-1] == val_a].sample(num_a)
    df_b = df[df.iloc[:,-1] == val_b].sample(num_b)
    
    #remove it
    df.drop(df_a.index, inplace=True)
    df.drop(df_b.index, inplace=True)
    
    # samples: df_a and df_b
    # simply combining two
    frames = [df_a, df_b]
    df_validation = pd.concat(frames)
    
    return df, df_validation

training, validation = partition_dataset(df)
```

```python
validation
```
